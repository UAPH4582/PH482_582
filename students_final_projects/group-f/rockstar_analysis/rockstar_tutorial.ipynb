# halo merger trees

The Rockstar halo finder treats each snapshot as independent, so halos at different snapshots do not know about each other. ConsistentTrees produces halo merger trees that link halos over time.

Note: ConsistentTrees applies some smoothing and physical consistency checks on halos over time, which leads to two important differences from the halo catalog: (1) not every halo in the catalog exists in the merger tree (especially those that are marginally resolved) (2) some halos in the merger tree are 'phantom' halos that have been interpolated across snapshots but do not exist in the halo catalog at that snapshot.

Like with the halo catalogs, the reader automatically looks if baryonic particle property files exists at each snapshot and will read and append these properties to the halo merger trees. Again, not all snapshots have baryonic particle properties, so not all halos in the tree will have valid values for these.

```python
# read halo merger trees across all snapshots
# this is a concatenated list of all halos across all snaphots, with pointers to progenitor and descendant halos

halt = rockstar.io.IO.read_tree()
```

```python
# halo merger tree properties are stored as dictionary

for k in halt.keys():
    print(k)
```

```python
# now each halo has a corredponding snapshot index

halt['snapshot']
```

```python
# so, get all halos at snapshot 600 (z = 0) and print their masses

hindices = np.where(halt['snapshot'] == 600)[0]
print(halt['mass'][hindices])
```

```python
# alternately, get catalog of halos only at snapshot 600 from tree

hal = rockstar.io.IO.get_catalog_from_tree(halt, 600)
```

```python
# example workflow

# start with a halo with tree index 0 (at z = 0)
hindex = 0
print(halt['snapshot'][hindex])
print(halt['mass'][hindex])

# find its progenitors, list their mass
prog_index = halt['progenitor.main.index'][hindex]
prog_indices = []
while prog_index >= 0:
    prog_indices.append(prog_index)
    prog_index = halt['progenitor.co.index'][prog_index]
print(prog_indices)
print(halt['mass'][prog_indices])

# for the main progenitor, find its progenitors, list their mass
hindex = prog_indices[0]
prog_index = halt['progenitor.main.index'][hindex]
prog_indices = []
while prog_index >= 0:
    prog_indices.append(prog_index)
    prog_index = halt['progenitor.co.index'][prog_index]
print(prog_indices)
print(halt['mass'][prog_indices])
```

```python
# get list of main progenitors as far back as can go

hindex = 0
prog_main_index = hindex
prog_main_indices = []
while prog_main_index >= 0:
    prog_main_indices.append(prog_main_index)
    prog_main_index = halt['progenitor.main.index'][prog_main_index]

print(prog_main_indices)
print(halt['mass'][prog_main_indices])
print(halt['position'][prog_main_indices])
```

```python
# halt.prop() computes derived quantities and can make walking the tree easier

# for halo with tree index = 0, get its main progenitors going back as far as can (including self)
hindex = 0
prog_indices = halt.prop('progenitor.main.indices', hindex)

# print mass history
print(prog_indices)
print(halt['mass'][prog_indices])
```

```python
# print stellar mass history - recall that not all snapshots necessarily have star particle information

print(halt['star.mass'][prog_indices])
```

```python
# for halo with tree index = 1, get all of its progenitors at previous snapshot, print their masses

hindex = 1
prog_indices = halt.prop('progenitor.indices', hindex)

print(prog_indices)
print(halt['mass'][prog_indices])
```

```python
# for halo with tree index = 123456, get its descendants going forward as far as can (including self)

hindex = 123456
desc_indices = halt.prop('descendant.indices', hindex)

# print future mass growth
print(desc_indices)
print(halt['mass'][desc_indices])
```

# interfacting between halo catalog and merger trees

Each halo in the merger trees has a pointers to its array index in the halo catalog: 'catalog.index'

Conversely, each halo in the catalog has a pointer to its array index in the halo merger trees: 'tree.index'

You can use these to go back and forth between the two.

Note: tree ids/indices are unique across all snapshots, but halo catalog ids/indices are unique only at a given snapshot

```python
# read halo merger trees across all snapshots
halt = rockstar.io.IO.read_tree()

# read halo catalog at each snapshot
hals = rockstar.io.IO.read_catalogs(snapshot_values='all')
```

```python
# starting from the halo catalog at z = 0, get index of the halo in the merger trees
hal = hals[600]

cat_index = 100  # random halo of interest
print(hal['mass'][cat_index])
tree_index = hal['tree.index'][cat_index]
print(tree_index)
print(halt['mass'][tree_index])
```

```python
# conversely, starting from the halo merger trees, get index of the halo in the snapshot catalog
# because halo catalog indices are *not* unique across snapshots, 
# you also need to use which snapshot index the halo points to

tree_index = 100  # random halo of interest
print(halt['mass'][tree_index])
snapshot_index = halt['snapshot'][tree_index]  # index of snapshot in halo catalog
cat_index = halt['catalog.index'][tree_index]  # halo catalog index at snapshot
print(snapshot_index, cat_index)
print(hals[snapshot_index]['mass'][cat_index])
```
